```python
def get_edge_coords(stomach,ref_stomach,ref_bounds,back=False):
    '''This function return the indices of the back and top tunnel edges.
    stomach_element: needs to be a gdsCAD element'''
    
    if back:
        back_ind1 = np.where((ref_stomach.points[:,0]==ref_bounds[1][0][0])&(ref_stomach.points[:,1]==ref_bounds[1][0][1]))[0][0]
        back_ind2 = np.where((ref_stomach.points[:,0]==ref_bounds[0][-1][0])&(ref_stomach.points[:,1]==ref_bounds[0][-1][1]))[0][0]
        p1 = stomach.points[back_ind2].copy() #Imprtant to copy(), otherwise instance is passed!!
        p2 = stomach.points[back_ind1].copy()
    else: 
        top_ind1 = np.where((ref_stomach.points[:,0]==ref_bounds[1][-1][0])&(ref_stomach.points[:,1]==ref_bounds[1][-1][1]))[0][0]
        top_ind2 = np.where((ref_stomach.points[:,0]==ref_bounds[0][0][0])&(ref_stomach.points[:,1]==ref_bounds[0][0][1]))[0][0]
        p1 = stomach.points[top_ind2].copy()
        p2 = stomach.points[top_ind1].copy()
    return p2,p1
```

```python

def make_bezier(xys):
    # xys should be a sequence of 2-tuples (Bezier control points)
    n = len(xys)
    combinations = pascal_row(n-1)
    def bezier(ts):
        # This uses the generalized formula for bezier curves
        # http://en.wikipedia.org/wiki/B%C3%A9zier_curve#Generalization
        result = []
        for t in ts:
            tpowers = (t**i for i in range(n))
            upowers = reversed([(1-t)**i for i in range(n)])
            coefs = [c*a*b for c, a, b in zip(combinations, tpowers, upowers)]
            result.append(
                tuple(sum([coef*p for coef, p in zip(coefs, ps)]) for ps in zip(*xys)))
        return result
    return bezier

def pascal_row(n):
    # This returns the nth row of Pascal's Triangle
    result = [1]
    x, numerator = 1, n
    for denominator in range(1, n//2+1):
        # print(numerator,denominator,x)
        x *= numerator
        x /= denominator
        result.append(x)
        numerator -= 1
    if n&1 == 0:
        # n is even
        result.extend(reversed(result[:-1]))
    else:
        result.extend(reversed(result)) 
    return result
```

```python
#Roll outershape such that first point is close to first point of inner shape
outer_shape = np.roll(np.concatenate((relevant_bounds[0],relevant_bounds[1]),axis=0),-2760,axis=0)
outer_shape = np.concatenate((outer_shape,[outer_shape[0]]),axis=0).tolist()
inner_shape = relevant_bounds[2].tolist()
points = inner_shape+outer_shape

stomach_1 = g.core.Boundary(points,layer=1)
stomach_2 = stomach_1.copy().translate((-4000,0))
stomach_2_5 = stomach_1.copy().reflect('x').translate((-4000,0))
stomach_3 = stomach_1.copy().rotate(270).translate((-5670,2300))
stomach_4 = stomach_1.copy().rotate(90).reflect('y').translate((-5670,-2300))

entr_1 = g.shapes.Disk((0,0),350,layer=2)
entr_2 = entr_1.copy().translate((-4000,0))
entr_3 = entr_1.copy().translate((-5670,2300))
entr_4 = entr_1.copy().translate((-5670,-2300))

#Create stright channel between stomach_1 and stomach_2
top_1,top_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds)
back_1,back_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds,back=True)
tunnel = np.stack((back_1,back_2,top_2,top_1))
channel_1_2 = g.core.Boundary(tunnel,layer=1)


ts = [t/100.0 for t in range(101)]
#Straight tunnel between stomach 2 and stomach 3
top_1,top_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds)

top_1[0]+=0
top_1[1]+=-58
top_2[0]+=-58
top_2[1]+=0

back_1,back_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds,back=True)

tunnel = np.concatenate((top_2,top_1,back_1,back_2),axis=0).reshape((4,2))

channel_2_3 = g.core.Boundary(tunnel,layer=1)


#Straight tunnel between stomach 2 and stomach 4

#top_1,top_2 chan use these from first tunnel
back_2,back_1 = get_edge_coords(stomach_4,stomach_1,relevant_bounds,back=True)

tunnel = np.concatenate((top_2,top_1,back_2,back_1),axis=0).reshape((4,2))

channel_2_4 = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_3
top_1,top_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]+500])
end_2 = np.array([top_2[0],top_2[1]+500])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_3_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_4
top_1,top_2 = get_edge_coords(stomach_4,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]-500])
end_2 = np.array([top_2[0],top_2[1]-500])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_4_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_1
back_1,back_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds,back=True)
end_1 = np.array([back_1[0]+500,back_1[1]])
end_2 = np.array([back_2[0]+500,back_2[1]])
tunnel = np.stack((back_1,back_2,end_2,end_1))
channel_1_end = g.core.Boundary(tunnel,layer=1)


#Group elements of layer 1 together
stomach_group = g.core.Elements((stomach_1,stomach_2,stomach_2_5,stomach_3,stomach_4,channel_1_2,channel_2_3,channel_2_4,channel_3_end,channel_4_end,channel_1_end))
entr_group = g.core.Elements((entr_1,entr_2,entr_3,entr_4))

#Create cell
network_ortho = g.core.Cell('NETWORK')
network_ortho.add([stomach_group,entr_group])

plt.figure(figsize=(25,15))
network_ortho.show()
```

```python
outer_shape = np.roll(np.concatenate((relevant_bounds[0],relevant_bounds[1]),axis=0),-2760,axis=0)
outer_shape = np.concatenate((outer_shape,[outer_shape[0]]),axis=0).tolist()
inner_shape = relevant_bounds[2].tolist()
points = inner_shape+outer_shape

stomach_1 = g.core.Boundary(points,layer=1)
stomach_2 = stomach_1.copy().translate((-4000,0))
stomach_2_5 = stomach_1.copy().reflect('x').translate((-4000,0))
stomach_3 = stomach_1.copy().rotate(270).translate((-7500,2300))
stomach_4 = stomach_1.copy().rotate(90).reflect('y').translate((-7500,-2300))

entr_1 = g.shapes.Disk((0,0),350,layer=2)
entr_2 = entr_1.copy().translate((-4000,0))
entr_3 = entr_1.copy().translate((-7500,2300))
entr_4 = entr_1.copy().translate((-7500,-2300))

#Create stright channel between stomach_1 and stomach_2
top_1,top_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds)
back_1,back_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds,back=True)
tunnel = np.stack((back_1,back_2,top_2,top_1))
channel_1_2 = g.core.Boundary(tunnel,layer=1)


ts = [t/100.0 for t in range(101)]
#First Bezier channel
top_1,top_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds)

top_1[0]+=1000
top_1[1]+=90
top_2[0]+=1000
top_2[1]+=-90

back_1,back_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds,back=True)

xys = [back_1, (back_1[0], top_1[1]), top_1]
bezier = make_bezier(xys)
points_1 = bezier(ts)

xys = [back_2, (back_2[0], top_2[1]), top_2]
bezier = make_bezier(xys)
points_2 = bezier(ts)

tunnel = np.append(points_1,np.flipud(points_2),axis=0)
channel_2_3 = g.core.Boundary(tunnel,layer=1)


#Second Bezier channel

#top_1,top_2 chan use these from first bezier channel
back_2,back_1 = get_edge_coords(stomach_4,stomach_1,relevant_bounds,back=True)

xys = [back_1, (back_1[0], top_1[1]), top_1]
bezier = make_bezier(xys)
points_1 = bezier(ts)

xys = [back_2, (back_2[0], top_2[1]), top_2]
bezier = make_bezier(xys)
points_2 = bezier(ts)

tunnel = np.append(points_1,np.flipud(points_2),axis=0)
channel_2_4 = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_3
top_1,top_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]+500])
end_2 = np.array([top_2[0],top_2[1]+500])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_3_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_4
top_1,top_2 = get_edge_coords(stomach_4,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]-500])
end_2 = np.array([top_2[0],top_2[1]-500])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_4_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_1
back_1,back_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds,back=True)
end_1 = np.array([back_1[0]+500,back_1[1]])
end_2 = np.array([back_2[0]+500,back_2[1]])
tunnel = np.stack((back_1,back_2,end_2,end_1))
channel_1_end = g.core.Boundary(tunnel,layer=1)


#Group elements of layer 1 together
stomach_group = g.core.Elements((stomach_1,stomach_2,stomach_2_5,stomach_3,stomach_4,channel_1_2,channel_2_3,channel_2_4,channel_3_end,channel_4_end,channel_1_end))
entr_group = g.core.Elements((entr_1,entr_2,entr_3,entr_4))

#Create cell
network_bezier = g.core.Cell('NETWORK')
network_bezier.add([stomach_group,entr_group])

plt.figure(figsize=(25,15))
network_bezier.show()
```
