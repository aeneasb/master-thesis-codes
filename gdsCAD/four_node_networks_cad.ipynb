```python
def get_edge_coords(stomach,ref_stomach,ref_bounds,back=False):
    '''This function return the indices of the back and top tunnel edges.
    stomach_element: needs to be a gdsCAD element'''
    
    if back:
        top_ind1 = np.where((ref_stomach.obj[0].points[:,0]==ref_bounds[1][-1][0])&(ref_stomach.obj[0].points[:,1]==ref_bounds[1][-1][1]))[0][0]
        top_ind2 = np.where((ref_stomach.obj[0].points[:,0]==ref_bounds[1][0][0])&(ref_stomach.obj[0].points[:,1]==ref_bounds[1][0][1]))[0][0]
        p1 = stomach.obj[0].points[top_ind2].copy() #Important to copy(), otherwise instance is passed!!
        p2 = stomach.obj[0].points[top_ind1].copy()
    else: 
        top_ind1 = np.where((ref_stomach.obj[1].points[:,0]==ref_bounds[2][-1][0])&(ref_stomach.obj[1].points[:,1]==ref_bounds[2][-1][1]))[0][0]
        top_ind2 = np.where((ref_stomach.obj[1].points[:,0]==ref_bounds[2][0][0])&(ref_stomach.obj[1].points[:,1]==ref_bounds[2][0][1]))[0][0]
        p1 = stomach.obj[1].points[top_ind2].copy()
        p2 = stomach.obj[1].points[top_ind1].copy()
    return p2,p1
```

```python
def make_bezier(xys):
    # xys should be a sequence of 2-tuples (Bezier control points)
    n = len(xys)
    combinations = pascal_row(n-1)
    def bezier(ts):
        # This uses the generalized formula for bezier curves
        # http://en.wikipedia.org/wiki/B%C3%A9zier_curve#Generalization
        result = []
        for t in ts:
            tpowers = (t**i for i in range(n))
            upowers = reversed([(1-t)**i for i in range(n)])
            coefs = [c*a*b for c, a, b in zip(combinations, tpowers, upowers)]
            result.append(
                tuple(sum([coef*p for coef, p in zip(coefs, ps)]) for ps in zip(*xys)))
        return result
    return bezier

def pascal_row(n):
    # This returns the nth row of Pascal's Triangle
    result = [1]
    x, numerator = 1, n
    for denominator in range(1, n//2+1):
        # print(numerator,denominator,x)
        x *= numerator
        x /= denominator
        result.append(x)
        numerator -= 1
    if n&1 == 0:
        # n is even
        result.extend(reversed(result[:-1]))
    else:
        result.extend(reversed(result)) 
    return result
```

```python
#building blocks
tail = g.core.Boundary(np.concatenate((relevant_bounds[1],[relevant_bounds[1][0]]),axis=0),layer=1)
bulk = g.core.Boundary(np.concatenate((relevant_bounds[2],[relevant_bounds[2][0]]),axis=0),layer=1)
arch = g.core.Boundary(relevant_bounds[0],layer=1)

stomach_1 = g.core.Elements((tail,bulk,arch))
stomach_2 = stomach_1.copy().translate((-650,0))
stomach_2_5 = stomach_1.copy().reflect('x').translate((-650,0))

top_1,top_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds)
trans = top_1[0]-diff/2

stomach_3 = stomach_1.copy().rotate(270).translate((trans,380)).reflect('y',origin=(trans,380))
stomach_4 = stomach_1.copy().rotate(90).reflect('y').translate((trans,-380)).reflect('y',origin=(trans,-380))

entr_1 = g.shapes.Disk((0,0),58,layer=2)
entr_2 = entr_1.copy().translate((-650,0))
entr_3 = entr_1.copy().translate((trans,380))
entr_4 = entr_1.copy().translate((trans,-380))

#Create stright channel between stomach_1 and stomach_2
top_1,top_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds)
back_1,back_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds,back=True)
tunnel = np.stack((back_1,back_2,top_1,top_2))
channel_1_2 = g.core.Boundary(tunnel,layer=1)

#Straight tunnel between stomach 2 and stomach 3
top_1,top_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds)

top_1_prime = np.array((top_1[0]-diff+0.05,top_1[1]))
top_2_prime = np.array((top_2[0]-diff+0.05,top_2[1]))

back_1,back_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds,back=True)

tunnel = np.concatenate((top_1_prime,top_1,back_2,back_1),axis=0).reshape((4,2))

channel_2_3 = g.core.Boundary(tunnel,layer=1)


#Straight tunnel between stomach 2 and stomach 4

#top_1,top_2 chan use these from first tunnel
back_2,back_1 = get_edge_coords(stomach_4,stomach_1,relevant_bounds,back=True)

tunnel = np.concatenate((top_2_prime,top_2,back_1,back_2),axis=0).reshape((4,2))

channel_2_4 = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_3
top_1,top_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]+80])
end_2 = np.array([top_2[0],top_2[1]+80])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_3_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_4
top_1,top_2 = get_edge_coords(stomach_4,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]-80])
end_2 = np.array([top_2[0],top_2[1]-80])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_4_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_1
back_1,back_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds,back=True)
end_1 = np.array([back_1[0]+80,back_1[1]])
end_2 = np.array([back_2[0]+80,back_2[1]])
tunnel = np.stack((back_1,back_2,end_2,end_1))
channel_1_end = g.core.Boundary(tunnel,layer=1)


#Group elements of layer 1 together
channel_group = g.core.Elements((channel_1_2,channel_2_3,channel_2_4,channel_3_end,channel_4_end,channel_1_end))
entr_group = g.core.Elements((entr_1,entr_2,entr_3,entr_4))

#Create cell
network_ortho = g.core.Cell('NETWORK')
network_ortho.add([stomach_1,stomach_2,stomach_2_5,stomach_3,stomach_4,entr_group,channel_group])

plt.figure(figsize=(20,15))
objects = network_ortho.show()
```

```python
#building blocks
tail = g.core.Boundary(np.concatenate((relevant_bounds[1],[relevant_bounds[1][0]]),axis=0),layer=1)
bulk = g.core.Boundary(np.concatenate((relevant_bounds[2],[relevant_bounds[2][0]]),axis=0),layer=1)
arch = g.core.Boundary(relevant_bounds[0],layer=1)

stomach_1 = g.core.Elements((tail,bulk,arch))
stomach_3 = stomach_1.copy().rotate(270).translate((-650,500)).reflect('y',origin=(-650,500))
stomach_4 = stomach_1.copy().rotate(90).reflect('y').translate((-650,-500)).reflect('y',origin=(-650,-500))

bulk_var = g.core.Boundary(np.concatenate((bulk_variant,[bulk_variant[0]]),axis=0),layer=1)
arch_var = g.core.Boundary(np.concatenate((arch_variant,[arch_variant[0]]),axis=0),layer=1)
stomach_2 = g.core.Elements((tail.copy(),bulk_var,arch_var)).translate((-650,0))
stomach_2_5 = stomach_2.copy().reflect('x')

entr_1 = g.shapes.Disk((0,0),58,layer=2)
entr_2 = entr_1.copy().translate((-650,0))
entr_3 = entr_1.copy().translate((-650,500))
entr_4 = entr_1.copy().translate((-650,-500))

#Create stright channel between stomach_1 and stomach_2
top_1,top_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds)
back_1,back_2 = get_edge_coords(stomach_2,stomach_1,relevant_bounds,back=True)
tunnel = np.stack((back_1,back_2,top_1,top_2))
channel_1_2 = g.core.Boundary(tunnel,layer=1)


ts = [t/19.0 for t in range(20)]

#curvey tunnel between stomach 4 and stomach 3

back_3_1,back_3_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds,back=True)

back_4_1,back_4_2 = get_edge_coords(stomach_4,stomach_1,relevant_bounds,back=True)

xys = [back_3_2, (back_3_2[0], stomach_2.obj[1].points[6][1]), stomach_2.obj[1].points[6]]
bezier = make_bezier(xys)
points_right_up = bezier(ts)

xys = [stomach_2.obj[1].points[7], (stomach_2_5.obj[1].points[7][0]-40, 0), stomach_2_5.obj[1].points[7]]
bezier = make_bezier(xys)
points_straight_down = bezier(ts)

xys = [stomach_2_5.obj[1].points[7], (back_4_2[0], stomach_2_5.obj[1].points[7][1]), back_4_2]
bezier = make_bezier(xys)
points_right_bottom = bezier(ts)

xys = [back_4_1, (back_4_1[0]-15, stomach_2_5.obj[1].points[-1][1]), stomach_2_5.obj[1].points[-1]]
bezier = make_bezier(xys)
points_left_bottom = bezier(ts)

xys = [stomach_2.obj[1].points[-1], (back_3_1[0]-15, stomach_2.obj[1].points[-1][1]), back_3_1]
bezier = make_bezier(xys)
points_left_up = bezier(ts)

tunnel_complex = np.concatenate((points_right_up,points_straight_down,points_right_bottom,points_left_bottom,points_left_up),axis=0)
tunnel_complex = tunnel_complex.reshape((len(points_right_up)*5,2)).tolist()
channel_3_4 = g.core.Boundary(tunnel_complex,layer=1)

#Create stright end channel of stomach_3
top_1,top_2 = get_edge_coords(stomach_3,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]+80])
end_2 = np.array([top_2[0],top_2[1]+80])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_3_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_4
top_1,top_2 = get_edge_coords(stomach_4,stomach_1,relevant_bounds)
end_1 = np.array([top_1[0],top_1[1]-80])
end_2 = np.array([top_2[0],top_2[1]-80])
tunnel = np.stack((top_2,top_1,end_1,end_2))
channel_4_end = g.core.Boundary(tunnel,layer=1)

#Create stright end channel of stomach_1
back_1,back_2 = get_edge_coords(stomach_1,stomach_1,relevant_bounds,back=True)
end_1 = np.array([back_1[0]+80,back_1[1]])
end_2 = np.array([back_2[0]+80,back_2[1]])
tunnel = np.stack((back_1,back_2,end_2,end_1))
channel_1_end = g.core.Boundary(tunnel,layer=1)

#Create Cap on top of stomach_2/stomach_2_5
xys = [stomach_2.obj[1].points[-1],(stomach_2.obj[1].points[-1][0]-70,0),stomach_2_5.obj[1].points[-1]]
bezier = make_bezier(xys)
cap_points = bezier(ts)
cap = g.core.Boundary(cap_points,layer=1)

#Group elements of layer 1 together
channel_group = g.core.Elements((cap,channel_1_2,channel_3_4,channel_3_end,channel_4_end,channel_1_end))
entr_group = g.core.Elements((entr_1,entr_2,entr_3,entr_4))

#Create cell
network_bulk = g.core.Cell('NETWORK')
network_bulk.add([stomach_1,stomach_2,stomach_2_5,stomach_3,stomach_4,channel_group,entr_group])

plt.figure(figsize=(25,15))
network_bulk.show()
```
