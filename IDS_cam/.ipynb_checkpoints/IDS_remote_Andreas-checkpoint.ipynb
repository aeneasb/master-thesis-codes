```python
class Camera(ids.Camera):
    def __init__(self,color_mode=ids.ids_core.COLOR_RGB8):
        
        print('  >> power cycling USB ports')
        self.hard_reset()
        time.sleep(1)
        
        print('  >> connecting to camera')
        super(Camera,self).__init__()
        
        print('  >> initializing camera settings')
        self.color_mode = color_mode # Get images in RGB format
        
        self.auto_exposure = False
        print('  >> Done camera initialization')
        
    def getImage(self,startTime = 0):
    """gets a new image and saves it in the variable self. img."""

    self.attempts = 0

    image,retval = self.get_raw_img()
    
    # if could not capture image
    while retval == False:
        self.attempts += 1
        if self.attempts == 100:
            print('Reconnect camera')
            time.sleep(2)
            self.reconnectCamera()

        if self.attempts > 1000:
            print('Unable to capture image')
            return 0, 0

        image, retval = self.get_raw_img()

    self.acquisitionTime = time.time() - startTime
    self.img = image

    return image, retval

    def get_raw_img(self):
        """
        Capture an image and return an array containing the data in raw 12-bit Bayer format
        :param labjack: optional , is required to do 

        """          
        try:
            # turn camera on/off before/after capturing image
                                   
            # get the next image in the buffer.

            self.continuous_capture = True
            data,meta = self.next()         
            self.continuous_capture = False

            # The camera returns the image in 16-bit data but the first 4 bits are all zero            
            data = data.astype('uint16')
            data = np.right_shift(data,4)
        
            # Rotate and flip the image for more natural display
            #data = np.rot90(data)
            #data = np.rot90(data)
            #data = np.fliplr(data)


            # if capture was successful, return True
            retval = True

        except Exception:

            # If unable to capture image, return False and no image
            print('Unable to capture image, try again')
            retval = False
            data   = None
        return data, retval
    
    def save_img(self, name, img, metadata = None):
        """
        Save a given image as a .tiff
        """
        img, retval = self.get_raw_img()
        # try:
        image_name = name + '.tiff'

        while os.path.isfile(image_name):
            try:
                imgNo = int(name.split('_')[-1])
                name = name.replace(str(imgNo),str(imgNo+1))
                imgNo += 1
                image_name = name + '.tiff'
                
            except:
                imgNo = 1
                name = name.replace(str(imgNo),str(imgNo+1))
                imgNo += 1
                image_name = name + '.tiff'


        img = np.array(img, dtype='uint16')
        img = Image.fromarray(img,'I;16')

        if metadata != None:
            img.save(name, 'TIFF', description = str(metadata))

        else:
            img.save(name,'TIFF')
            
    def hard_reset(self):
        num_ports = 2

        # power cycle USB ports
        for i in range(num_ports):
            # disable ASLR
            os.system('echo 0 | sudo tee /sys/bus/usb/devices/usb' + str(i+1) + '/power/autosuspend')
            
    def reconnectCamera(self):
        try:
            # import ids module
            from Reader_Control.Camera_Control.Camera import Camera
            import ids

            # Free all allocated memory for storing images.
            self.main_widget.setup.camera.free_all()
            # Closes open camera
            self.main_widget.setup.camera.close()

            # seting up camera
            print('> Initializing camera')
            self.main_widget.setup.camera =  Camera(
                    starting_exposure = 60,
                    color_mode = ids.ids_core.COLOR_BAYER_16,
                    magnification = 20
                )
        except:
            print('  >> could not connect to camera. Set offline mode.')
            self.main_widget.offline_mode = True
```
