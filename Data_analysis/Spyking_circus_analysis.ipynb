```python
#Filter the cropped traces
fs = 20000
lowcut = 300
highcut = 9500
order = 2

nyq = 0.5*fs

low = lowcut/nyq
high = highcut/nyq
b,a = butter(order,[low,high],btype='band',analog=False)

cut_data_butter = np.empty_like(cropped_data)
cut_data_butter = filtfilt(b, a, cropped_data)
```

```python
y_offset = (filtdat.metadata['y_clean'] - np.min(filtdat.metadata['y_clean']))+1
x_offset = (filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean']))+1

newpath=('/').join(path.split('/')[:-1])

fig=plt.figure(figsize=(17,3.5))
ax=fig.add_subplot(111)
ax.set_xlim(0,np.max(x_offset*17.5)+17.5)
ax.set_ylim(0,np.max(y_offset*17.5)+17.5)
ax.set_aspect('equal')
ax.set_facecolor('k')
ax.set_ylabel(r'$\mu m$',fontsize=18)
ax.set_xlabel(r'$\mu m$',fontsize=18)
ax.tick_params(axis='both',labelsize=14)

value = results['spiketimes']['temp_1'][:]
extent_trace=8 #Trace extend on the canvas
bin_size = 90
x_trace=np.linspace(-extent_trace,extent_trace,bin_size) #Trace extend in datapoints
index = np.asarray([np.arange(-int(bin_size*(1/3.)),int(bin_size*(2/3.)))+c for c in value]).flatten()

for idx,s in enumerate(np.delete(filtdat.butter_data,115,axis=0)):
    trace_stack = s[index].reshape((len(value),bin_size)) #Extract all bins where spikes have been fitted.
    std_trace = 0.5*np.std(trace_stack,axis=0)
    mean_trace = 0.5*np.mean(trace_stack,axis=0)

    ax.fill_between(x_trace+np.delete(x_offset,115)[idx]*17.5,np.delete(y_offset,115)[idx]*17.5\
                    +mean_trace+std_trace,np.delete(y_offset,115)[idx]*17.5+mean_trace-std_trace,color='w',linewidth=0.1,alpha=0.5)
    ax.plot(x_trace+np.delete(x_offset,115)[idx]*17.5,np.delete(y_offset,115)[idx]*17.5+mean_trace,linewidth=0.6,color='r')

bar = AnchoredSizeBar(ax.transData,
                           16, r'$3 ms$',3, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0)
ax.add_artist(bar)

bar_rotation = transforms.Affine2D().rotate_deg(90) #Rotate the coordinate system
abs_bar_rotation = ax.transData + bar_rotation
bar_ver = AnchoredSizeBar(abs_bar_rotation,
                           14, r'$660 \mu V$',1, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0,
                            label_top=False,
                             borderpad=0,sep=1)

ax.add_artist(bar_ver)
plt.savefig(newpath+'/temp_1.png',dpi=400,bbox_inches='tight',pad_inches=0)
```
