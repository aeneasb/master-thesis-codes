```python
y_offset = (filtdat.metadata['y_clean'] - np.min(filtdat.metadata['y_clean']))+1
x_offset = (filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean']))+1


newpath=('/').join(path.split('/')[:-1])

fig=plt.figure(figsize=(17,3.5))
ax=fig.add_subplot(111)
ax.set_xlim(0,np.max(x_offset*17.5)+17.5)
ax.set_ylim(0,np.max(y_offset*17.5)+17.5)
ax.set_aspect('equal')
ax.set_facecolor('k')
ax.set_ylabel(r'$\mu m$',fontsize=18)
ax.set_xlabel(r'$\mu m$',fontsize=18)
ax.tick_params(axis='both',labelsize=14)
extent_trace=8 #+size_square*17.5#if plotting all electrodes. 


for ids,s in enumerate(neg_spikes):
    trace_stack = np.zeros([60])
    for i in s[1]:
        trace_stack = np.vstack((trace_stack, filtdat.butter_data[s[0],i-30:i+30]))

    x_trace=np.linspace(-extent_trace,extent_trace,trace_stack.shape[1])
    if trace_stack.shape[0]>2:
        std_trace = 0.5*np.std(trace_stack[1:,:],axis=0)
        mean_trace = 0.5*np.mean(trace_stack[1:,:],axis=0)
        ax.fill_between(x_trace+x_offset[ids]*17.5,y_offset[ids]*17.5+mean_trace+2*std_trace,\
                        y_offset[ids]*17.5+mean_trace-std_trace,color='w',linewidth=0.1,alpha=0.6)
        ax.plot(x_trace+x_offset[ids]*17.5,y_offset[ids]*17.5+mean_trace,linewidth=0.6,color='r')
    else:
        pass
        #ax.plot(x_trace+x_offset[ids]*17.5,y_offset[ids]*17.5+trace_stack[1,:],alpha=0.3,linewidth=0.6,color='r')

label_font_prop = FontProperties(size=16)
bar = AnchoredSizeBar(ax.transData,
                           16, r'$3 ms$',3, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0,fontproperties=label_font_prop)
ax.add_artist(bar)

bar_rotation = transforms.Affine2D().rotate_deg(90) #Rotate the coordinate system
abs_bar_rotation = ax.transData + bar_rotation
bar_ver = AnchoredSizeBar(abs_bar_rotation,
                           14, r'$660 \mu V$',1, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0,
                            label_top=False,
                            borderpad=0,sep=1,fontproperties=label_font_prop)

ax.add_artist(bar_ver)


plt.savefig(newpath+'/conf_mean_neg_hor.png',dpi=400,bbox_inches='tight',pad_inches=0)
```

##  Convolve the spike trains

```python
pos_spikes = [i for i in enumerate(pos_ind) if i[1]]
neg_spikes = [i for i in enumerate(neg_ind ) if i[1]]
print '{0:d} electrodes have positive spikes, {1:d} electrodes have negative spikes'.format(len(pos_spikes),len(neg_spikes))
```

```python
#Convolve the spike trains
size = 147
similarity_matrix=np.zeros((size,size))

for i in range(size):
    for j in range(i+1,size):
        if neg_ind[i] and neg_ind[j]:
            convo=fftconvolve(neg_ind[i],neg_ind[j])
            similarity_matrix[i,j]=np.max(convo)
        else:
            similarity_matrix[i,j]=0
```

```python
#And plot them
similarity_matrix=similarity_matrix+similarity_matrix.T

fig = plt.figure(figsize=(10,10))
ax=fig.add_axes([0,0,0.97,1])
ax.set_aspect('equal')
ax.set_xlabel(r'$spike train$',fontsize=40)
ax.set_ylabel(r'$spike train$',fontsize=40)
ax.set_facecolor('k')
ax.tick_params(axis='both',labelsize=30)

i = ax.imshow(similarity_matrix,cmap='hot',interpolation='none')
colorbar_ax = fig.add_axes([0.98,0,0.02,1])
cbar = fig.colorbar(i,cax=colorbar_ax)
cbar.ax.tick_params(labelsize=30)
cbar.ax.set_ylabel(r'$[conv_{max}]$',fontsize=40)
plt.savefig(path[:-31]+'/thesis_conv_matrix_spike_times.png',dpi=400,bbox_inches='tight',pad_inches=0)
```

## Plot critical traces separately

```python
x_offset = (filtdat.metadata['y_clean'] - np.min(filtdat.metadata['y_clean']))+1
y_offset = (filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean']))+1


newpath=('/').join(path.split('/')[:-1])

fig=plt.figure(figsize=(17,3.5))
ax=fig.add_subplot(111)
ax.set_xlim(0,np.max(x_offset*17.5)+17.5)
ax.set_ylim(0,np.max(y_offset*17.5)+17.5)
ax.set_aspect('equal')
ax.set_facecolor('k')
ax.set_ylabel(r'$\mu m$',fontsize=20)
ax.set_xlabel(r'$\mu m$',fontsize=20)
ax.tick_params(axis='both',labelsize=15)
extent_trace=8 #+size_square*17.5#if plotting all electrodes. 

x_offset_clean = np.delete(x_offset,filtdat.metadata['Critical_traces'],axis=0)
y_offset_clean = np.delete(y_offset,filtdat.metadata['Critical_traces'],axis=0)

clean = np.delete(filtdat.butter_data,filtdat.metadata['Critical_traces'],axis=0)
for ids,s in enumerate(neg_spikes):
    trace_stack = np.zeros([60])
    for i in s[1]:
        trace_stack = np.vstack((trace_stack, clean[s[0],i-30:i+30]))
    x_trace=np.linspace(-extent_trace,extent_trace,trace_stack.shape[1])

    if trace_stack.shape[0]>2:

        std_trace = np.std(trace_stack[1:,:],axis=0)
        mean_trace = np.mean(trace_stack[1:,:],axis=0)
        ax.fill_between(x_trace+x_offset_clean[ids]*17.5,y_offset_clean[ids]*17.5+mean_trace+2*std_trace,\
                        y_offset_clean[ids]*17.5+mean_trace-std_trace,color='w',linewidth=0.1,alpha=0.4)
        ax.plot(x_trace+x_offset_clean[ids]*17.5,y_offset_clean[ids]*17.5+mean_trace,linewidth=0.4,color='r')
    else:
        ax.plot(x_trace+x_offset_clean[ids]*17.5,y_offset_clean[ids]*17.5+trace_stack[1,:],linewidth=0.4,color='r')

#Crit traces
crits = filtdat.butter_data[filtdat.metadata['Critical_traces']]
x_offset_crit = x_offset[filtdat.metadata['Critical_traces']]
y_offset_crit = y_offset[filtdat.metadata['Critical_traces']]        
for ids,s in enumerate(neg_spikes_crit):
    trace_stack = np.zeros([60])
    for i in s[1]:
        trace_stack = np.vstack((trace_stack,crits[s[0],i-30:i+30]))

    x_trace=np.linspace(-extent_trace,extent_trace,trace_stack.shape[1])
    if trace_stack.shape[0]>2:
        std_trace = np.std(trace_stack[1:,:],axis=0)
        mean_trace = np.mean(trace_stack[1:,:],axis=0)
        ax.fill_between(x_trace+x_offset_crit[ids]*17.5,y_offset_crit[ids]*17.5+mean_trace+2*std_trace,\
                        y_offset_crit[ids]*17.5+mean_trace-std_trace,color='w',linewidth=0.1,alpha=0.4)
        ax.plot(x_trace+x_offset_crit[ids]*17.5,y_offset_crit[ids]*17.5+mean_trace,linewidth=0.4,color='r')
    else:
        ax.plot(x_trace+x_offset_crit[ids]*17.5,y_offset_crit[ids]*17.5+trace_stack[1,:],linewidth=0.4,color='r')


bar = AnchoredSizeBar(ax.transData,
                           16, r'$3 ms$', 3, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0)
ax.add_artist(bar)
bar_rotation = transforms.Affine2D().rotate_deg(90) #Rotate the coordinate system
abs_bar_rotation = ax.transData + bar_rotation
bar_ver = AnchoredSizeBar(abs_bar_rotation,
                           14, r'$230 \mu V$',1, 
                           pad=0.1,
                           color='white',
                           frameon=False,
                           size_vertical=0,
                            label_top=False,
                             borderpad=0,sep=1)

ax.add_artist(bar_ver)
plt.savefig(newpath+'/conf_mean_neg_hor.png',dpi=600,bbox_inches='tight',pad_inches=0)
```
