## Similarity matrix by convolution

```python
#Takes long...
similarity_matrix=np.zeros((147,147))

for i in range(147):
    for j in range(i+1,147):
        convo=fftconvolve(filtdat.butter_data[i,:],filtdat.butter_data[j,:])
        similarity_matrix[i,j]=np.max(convo)    
```

```python
similarity_matrix=similarity_matrix+similarity_matrix.T

fig = plt.figure(figsize=(10,10))
ax=fig.add_axes([0,0,0.97,1])
ax.set_aspect('equal')
ax.set_xlabel(r'$electrode$',fontsize=15)
ax.set_ylabel(r'$electrode$',fontsize=15)
ax.set_facecolor('k')
ax.tick_params(axis='both',labelsize=10)

i = ax.imshow(similarity_matrix,cmap='hot',interpolation='none')
colorbar_ax = fig.add_axes([0.98,0,0.02,1])
cbar = fig.colorbar(i,cax=colorbar_ax)
cbar.ax.tick_params(labelsize=20)
cbar.ax.set_ylabel(r'$[conv_{max}]$',fontsize=25)
plt.savefig(path[:-31]+'/thesis_conv_matrix.png',dpi=200,bbox_inches='tight',pad_inches=0)
```

## Similarity matrix by crosscorrelation

```python
similarity_matrix_correlate=np.zeros((147,147))

for i in range(147):
    for j in range(i+1,147):
        convo=np.correlate(filtdat.butter_data[i,:],filtdat.butter_data[j,:])
        similarity_matrix_correlate[i,j]=np.max(convo)
```

```python
plt.figure(figsize=(10,10))
similarity_matrix=similarity_matrix_correlate+similarity_matrix_correlate.T
plt.imshow(similarity_matrix,'hot')
```

```python
#Let the user choose an electrode
%matplotlib
xses=[]
yses=[]
indices = []
for i in range(1):
    fig=plt.figure(figsize=(22,12))
    imgplot = plt.imshow(filtdat.metadata['electrode_map'],'hot')
    points = np.around(np.asarray(plt.ginput(n=200,show_clicks=True,mouse_add=1,mouse_pop=3,mouse_stop=2)))
    plt.close()
    xses.append(map(int,[t[0] for t in points]))
    yses.append(map(int,[t[1] for t in points]))
    indices.append(xses[i][i]+yses[i][i]*220)
    
print 'The absolute index of the electrode is {}'.format(indices)
%matplotlib inline
```

## Compute the louvain community

```python
G=nx.from_numpy_matrix(similarity_matrix)
```

```python
partition = community.best_partition(G)
```

```python
from itertools import groupby
```

```python
groups = list(partition.viewvalues())
counts = [len(list(group)) for key, group in groupby(sorted(groups))]
print counts
```

```python
num_gr = 2
values = [counts.index(sorted(counts)[-i]) for i in range(1,num_gr+1)]
```

```python
y_offset = (filtdat.metadata['y_clean'] - np.min(filtdat.metadata['y_clean']))
x_offset = (filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean']))
span_x = np.max(filtdat.metadata['x_clean'])-np.min(filtdat.metadata['x_clean'])
span_y = np.max(filtdat.metadata['y_clean'])-np.min(filtdat.metadata['y_clean'])
matrix = np.zeros((span_y+1,span_x+1))

for key,value in partition.iteritems():
    if value in values:
        matrix[y_offset[key],x_offset[key]] = value
fig=plt.figure(figsize=(11.5,2.5))
ax=fig.add_axes([0,0,0.98,1])
#ax.set_aspect('equal')
ax.set_xlabel(r'$electrode$',fontsize=15)
ax.set_ylabel(r'$electrode$',fontsize=15)
ax.set_facecolor('k')
ax.tick_params(axis='both',labelsize=10)
i = ax.imshow(matrix,cmap='hot',interpolation='none',origin='lower')
colorbar_ax = fig.add_axes([0.98,0,0.02,1])
cbar = fig.colorbar(i,cax=colorbar_ax)
cbar.ax.tick_params(labelsize=20)
cbar.ax.set_ylabel(r'$[corr]$',fontsize=25)
```

## Can the same correlations be detected on spike trains?

```python
#Spike detection: Single recording
pos_ind = sp.spiker(5,0.05,filtdat.butter_data,filtdat.metadata['time'],60,-60)
neg_ind = sp.spiker(5,0.05,-1*filtdat.butter_data,filtdat.metadata['time'],60,-60)

pos_spikes = [i for i in enumerate(pos_ind) if i[1]]
neg_spikes = [i for i in enumerate(neg_ind ) if i[1]]
print '{0:d} electrodes have positive spikes, {1:d} electrodes have negative spikes'.format(len(pos_spikes),len(neg_spikes))
```

```python
#Convolve the spike trains
similarity_matrix=np.zeros((141,141))

for i in range(141):
    for j in range(i+1,141):
        if neg_ind[i] and neg_ind[j]:
            convo=fftconvolve(neg_ind[i],neg_ind[j])
            similarity_matrix[i,j]=np.max(convo)
        else:
            similarity_matrix[i,j]=0
```

```python
#Convolve the spike trains
similarity_matrix=np.zeros((147,147))

for i in range(147):
    for j in range(i+1,147):
        if neg_ind[i] and neg_ind[j]:
            convo=fftconvolve(neg_ind[i],neg_ind[j])
            similarity_matrix[i,j]=np.max(convo)
        else:
            similarity_matrix[i,j]=0
```

```python
similarity_matrix=np.zeros((147,147))

for i in range(len(neg_spikes)):
    for j in range(i+1,len(neg_spikes)):
        convo=fftconvolve(neg_spikes[i][1],neg_spikes[j][1])
        similarity_matrix[i,j]=np.max(convo)
```

```python
#And plot them
similarity_matrix=similarity_matrix+similarity_matrix.T

fig = plt.figure(figsize=(10,10))
ax=fig.add_axes([0,0,0.97,1])
ax.set_aspect('equal')
ax.set_xlabel(r'$electrode$',fontsize=15)
ax.set_ylabel(r'$electrode$',fontsize=15)
ax.set_facecolor('k')
ax.tick_params(axis='both',labelsize=10)

i = ax.imshow(similarity_matrix,cmap='hot',interpolation='none')
colorbar_ax = fig.add_axes([0.98,0,0.02,1])
cbar = fig.colorbar(i,cax=colorbar_ax)
cbar.ax.tick_params(labelsize=20)
cbar.ax.set_ylabel(r'$[conv_{max}]$',fontsize=25)
#plt.savefig(path[:-31]+'/thesis_conv_matrix_spike_times.png',dpi=200,bbox_inches='tight',pad_inches=0)
```
