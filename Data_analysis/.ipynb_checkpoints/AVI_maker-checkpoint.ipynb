```python
raw = im.import_rawdata(path,0,'end',modified=False)
```

```python
filtdat = fm.Filtering(raw)
filtdat.butter_filter(300,3000)
filtdat.discard_out_of_bounds()
#filtdat.reshape(1000)
raw=None
```

```python
#Spike detection: Single recording
pos_ind = sp.spiker(5,0.05,filtdat.butter_data,filtdat.metadata['time'],60,-60)
neg_ind = sp.spiker(5,0.05,-1*filtdat.butter_data,filtdat.metadata['time'],60,-60)

pos_spikes = [i for i in enumerate(pos_ind) if i[1]]
neg_spikes = [i for i in enumerate(neg_ind ) if i[1]]
print '{0:d} electrodes have positive spikes, {1:d} electrodes have negative spikes'.format(len(pos_spikes),len(neg_spikes))
```

```python
#Calculate the maximal amplitude in each timestep over all traces in all recordings: For one h5 file
max_vals=[]
timestamp=np.arange(600,700,1) #every time entry
for t in range(len(timestamp)-1):
    max_vals.append(np.max(np.max(filtdat.butter_data[:,timestamp[t]:timestamp[t+1]],axis=1)-np.min(filtdat.butter_data[:,timestamp[t]:timestamp[t+1]],axis=1)))
maximum_timepiece_diffs=np.asarray(max_vals)

print np.mean(maximum_timepiece_diffs)
plt.plot(maximum_timepiece_diffs)
```

```python
offset_x = np.min(filtdat.metadata['x_clean'])
offset_y = np.min(filtdat.metadata['y_clean'])
span_x = np.max(filtdat.metadata['x_clean'])-np.min(filtdat.metadata['x_clean'])
span_y = np.max(filtdat.metadata['y_clean'])-np.min(filtdat.metadata['y_clean'])

matrix = np.zeros((span_y+1,span_x+1))
matrix[filtdat.metadata['y_clean']-np.min(filtdat.metadata['y_clean']),filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean'])]=max_vals

fig=plt.figure(figsize=(span_x/4,span_y/6)) #10/44
ax=fig.add_axes([0,0,0.95,1])
i = ax.imshow(matrix,cmap='hot',aspect='auto',vmin=0,origin='lower')
colorbar_ax = fig.add_axes([0.95,0,0.05,1])
ax.set_axis_off()
fig.colorbar(i,cax=colorbar_ax)
plt.savefig(path[:-34]+'/voltage_map.png',bbox_inches='tight',pad_inches=0)
plt.close()
```

```python
x_offset = (filtdat.metadata['y_clean'] - np.min(filtdat.metadata['y_clean']))+1
y_offset = (filtdat.metadata['x_clean']-np.min(filtdat.metadata['x_clean']))+1

newpath=('/').join(path.split('/')[:-1])

#fig=plt.figure(figsize=(17,3.5))
fig=plt.figure(figsize=(17,3.5))


extent_trace=8 #on figure
bin_size = 60
x_trace=np.linspace(-extent_trace,extent_trace,bin_size)
t_ms = 0
timestamp=np.arange(event-100,event+150,2)
for t in timestamp:
    ax=fig.add_subplot(111)
    ax.set_xlim(0,np.max(x_offset*17.5)+17.5)
    ax.set_ylim(0,np.max(y_offset*17.5)+17.5)
    ax.set_aspect('equal')
    ax.set_facecolor('k')
    ax.set_ylabel(r'$\mu m$',fontsize=18)
    ax.set_xlabel(r'$\mu m$',fontsize=18)
    ax.tick_params(axis='both',labelsize=14)
    bar = AnchoredSizeBar(ax.transData,16, r'$3 ms$', 3, pad=0.1,color='white',frameon=False,size_vertical=0)
    ax.add_artist(bar)
    bar_rotation = transforms.Affine2D().rotate_deg(90) #Rotate the coordinate system
    abs_bar_rotation = ax.transData + bar_rotation
    bar_ver = AnchoredSizeBar(abs_bar_rotation,14, r'$230 \mu V$',1, pad=0.1,color='white',frameon=False,size_vertical=0,label_top=False,borderpad=0,sep=1)
    ax.add_artist(bar_ver)
    if (t-timestamp[0])%20==0 and (t-timestamp[0])!=0: #change ms valu every 20th
        print t-timestamp[0]
        ax.text(0,110,'t='+str(t_ms)+'ms',color='w',fontsize=20)
        t_ms+=1
    else:
        ax.text(0,110,'t='+str(t_ms)+'ms',color='w',fontsize=20)
    for i,j in enumerate(filtdat.butter_data):
        ax.plot(x_trace+x_offset[i]*17.5,y_offset[i]*17.5+j[t-bin_size/2:t+bin_size/2],\
        linewidth=1,color='w')
    plt.savefig(newpath+'/event_1/'+str(t)+'.png',bbox_inches='tight',pad_inches=0)
    plt.cla()
        
```
