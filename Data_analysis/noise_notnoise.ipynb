### Aligned spikes generated by noise in the data must have neighbouring datapoint summing to zero

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
spike_train=np.random.normal(0,0.8,size=(100,100000)) #no spikes actually..
```

```python
positions=np.argmax(spike_train,axis=1)
positions
```

```python
window=100
x=np.arange(-window,window)
mean=[]
for i in range(len(spike_train)):

    plt.plot(x,spike_train[i,positions[i]-window:positions[i]+window],color='b',alpha=0.3)
    mean.append(spike_train[i,positions[i]-window:positions[i]+window])

    
```

```python
m=np.mean(mean,axis=0)
s=np.std(mean,axis=0)
plt.plot(x,np.mean(mean,axis=0))
plt.fill_between(x,m-s,m+s,alpha=0.5)
```
